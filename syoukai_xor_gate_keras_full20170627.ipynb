```python
#『詳解ディープラーニング』より引用のXORゲートの実装コード　keras版の解説メモ　自分用
#利用環境
#keras-2.0.5 numpy-1.13.0 scipy-0.19.1 theano-0.9.0
```

```python
#テストデータXORゲート作成
```

```python
import numpy as np
```

```python
X = np.array([[0,0],[0,1],[1,0],[1,1]])
Y = np.array([[0],[1],[1],[0]])
```

```python
X
```

```python
Y
```

```python
#XORゲートのデータラベル完成
```

```python
from keras.layers.core import Activation
from keras.layers.core import Dense
from keras.models import Sequential
from keras.optimizers import SGD
```

```python
model = Sequential()

model.add(Dense(input_dim=2, units=2)) #入力層〜隠れ層部分。unitsは省略可Dense(2,input_dim=2)でOK
model.add(Activation('sigmoid'))

model.add(Dense(units=1)) #隠れ層〜出力層部分。unitsは省略可Dense(1)でもOK
model.add(Activation('sigmoid'))

model.compile(loss='binary_crossentropy', optimizer=SGD(lr=0.1))
```

```python
#↓model.fit(X, Y, epochs=1500, batch_size=4)のepochsが大きいと（mac miniだと1000以上）固まることがある。
#model.fit(X, Y, epochs=500, batch_size=4)などとエポック数を少なめにセットして同じmodel.fitを繰り返すと
#学習済みオブジェクトは更新されるので、model.fit(X, Y, epochs=500, batch_size=4)＊4回とかにするとローカルテストには良い
```

```python
model.fit(X, Y, epochs=1500, batch_size=4)
```

```python
classes = model.predict_classes(X, batch_size=4)
probability = model.predict_proba(X, batch_size=4)
```

```python
print("classified:{}".format(classes==Y))
```

```python
print("probability:{}".format(probability))
```

```python
#上記のようにXORゲートの学習でも1500epochs程度では精度は不十分になることがわかる
```

```python
#と、いうわけで学習エポックを500追加してみる↓
```

```python
model.fit(X, Y, epochs=500, batch_size=4)
```

```python
classes = model.predict_classes(X, batch_size=4)
probability = model.predict_proba(X, batch_size=4)
print("classified:{}".format(classes==Y))
print("probability:{}".format(probability))
```

```python
#↑まだまだ精度が悪いのでさらに1000エポックを追加してみる↓
```

```python
model.fit(X, Y, epochs=1000, batch_size=4)
classes = model.predict_classes(X, batch_size=4)
probability = model.predict_proba(X, batch_size=4)
print("classified:{}".format(classes==Y))
print("probability:{}".format(probability))
```

```python
#↑3/4が正解になり中々精度が上がってきたので、さらに学習エポックを500追加↓
```

```python
model.fit(X, Y, epochs=500, batch_size=4)
classes = model.predict_classes(X, batch_size=4)
probability = model.predict_proba(X, batch_size=4)
print("classified:{}".format(classes==Y))
print("probability:{}".format(probability))
```

```python
#精度十分になった。
#ここまで1500＋500＋1000＋500＝3500エポックかかっている。
#本書には4000エポックが規定されている（mac miniだと4000エポックだとメモリオーバーになり固まった）ので、
#多層ニューラルネットワークの計算コストは高いことが示唆される。
```
